Median imputation: This is similar to mean imputation, but the missing values are replaced with the median of the non-missing values.
Mode imputation: This is the imputation technique that replaces missing values with the most frequent value in the dataset.
Random sample imputation: This is the imputation technique that replaces missing values with random values from the dataset.
Removal is the process of removing rows or columns that contain missing values. This is a more drastic approach than imputation, but it can be necessary if the amount of missing data is large.

## Q6. Using the wine quality data set, perform principal component analysis (PCA) to reduce the number of features. What is the minimum number of principal components required to explain 90% of the variance in the data?

Principal component analysis (PCA) is a statistical technique that can be used to reduce the number of features in a data set. PCA works by transforming the data into a new set of features called principal components. The principal components are ordered by the amount of variance that they explain. The first principal component explains the most variance, the second principal component explains the second most variance, and so on.

To perform PCA on the wine quality data set, we can use the following code:

```python
import pandas as pd
import numpy as np
from sklearn.decomposition import PCA

# Load the wine quality dataset
df = pd.read_csv('winequality-red.csv')

# Perform PCA
pca = PCA(n_components=11)
pca.fit(df)

# Print the explained variance ratio
explained_variance_ratio = pca.explained_variance_ratio_

# Print the minimum number of principal components required to explain 90% of the variance
number_of_components = np.argmax(np.cumsum(explained_variance_ratio) >= 0.90) + 1

print(number_of_components)
```

The output of the code shows that the minimum number of principal components required to explain 90% of the variance in the wine quality data set is 4.

This means that we can reduce the number of features in the wine quality data set from 11 to 4 without losing too much information. The four principal components that explain the most variance in the data set are:
