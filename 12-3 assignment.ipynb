## Q1. Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5 using Python. Interpret the results.

```python

from scipy import stats
import numpy as np

sample_mean = 50
sample_std = 5
n = 100 # sample size

# calculate the standard error
std_error = sample_std / np.sqrt(n)

# calculate the confidence interval
conf_int = stats.norm.interval(0.95, loc=sample_mean, scale=std_error)

print("The 95% confidence interval is:", conf_int)
```

## Q6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the area corresponding to a chi-square statistic of 15.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Define the degrees of freedom
df = 10

# Define the x-axis values
x = np.linspace(0, 30, 1000)

# Define the chi-square distribution
y = chi2.pdf(x, df)

# Plot the chi-square distribution
plt.plot(x, y, color='blue')

# Shade the area corresponding to a chi-square statistic of 15
x_shade = np.linspace(15, 30, 1000)
y_shade = chi2.pdf(x_shade, df)
plt.fill_between(x_shade, y_shade, color='blue', alpha=0.3)

# Label the axes
plt.xlabel('Chi-square Statistic')
plt.ylabel('Probability Density')
plt.title('Chi-Square Distribution with 10 Degrees of Freedom')

# Show the plot
plt.show()
```

## Q7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who prefer Coke.

confidence_interval = sample_proportion ± z_critical * standard_error
where:

sample_proportion is the proportion of people in the sample who prefer Coke

z_critical is the critical z-value at the 99% confidence level, which can be found using a z-table or the norm.ppf() function in Python

standard_error is the standard error of the sample proportion, which can be calculated as follows:

standard_error = sqrt((sample_proportion * (1 - sample_proportion)) / n)

where n is the sample size.

Using the given data, we have:

sample_proportion = 520/1000 = 0.52

n = 1000

z_critical = norm.ppf(0.995) (since we want the 99% confidence level and we're using a two-tailed test)

standard_error = sqrt((0.52 * 0.48) / 1000) = 0.022

We can now calculate the confidence interval:

confidence_interval = 0.52 ± 2.576 * 0.022

confidence_interval = (0.47, 0.57)

Therefore, we can say with 99% confidence that the true proportion of people in the population who prefer Coke is between 0.47 and 0.57.

## Q8. A researcher hypothesizes that a coin is biased towards tails. They flip the coin 100 times and observe 45 tails. Conduct a chi-square goodness of fit test to determine if the observed frequencies match the expected frequencies of a fair coin. Use a significance level of 0.05.

```python
from scipy import stats

obs_freq = [45, 55] # observed frequencies
exp_freq = [50, 50] # expected frequencies assuming a fair coin

# calculate the chi-square test statistic and p-value
chi2_stat, p_value = stats.chisquare(obs_freq, f_exp = exp_freq)

print("Chi-square test statistic:", chi2_stat)
print("P-value:", p_value)
```

## Q11. A random sample of 30 people was selected from a population with an unknown mean and standard deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a significance level of 0.05.

```python
from scipy import stats
import numpy as np

sample_mean = 72
sample_std = 10
n = 30

# calculate the t statistic
t_stat = (sample_mean - 70) / (sample_std / np.sqrt(n))

# calculate the degrees of freedom
df = n - 1

# calculate the p-value for a two-sided test
p_val = stats.t.sf(np.abs(t_stat), df) * 2

# print the results
print("t statistic:", t_stat)
print("p-value:", p_val)
```
