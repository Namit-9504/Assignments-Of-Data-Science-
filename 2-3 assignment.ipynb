### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.

Hunter in 2003 as a tool for visualizing scientific data. Matplotlib is widely used in data science, machine learning, and other fields where data visualization is critical.

Matplotlib is used for creating 2D plots, including line plots, scatter plots, bar plots, histograms, and pie charts. It can also be used to create more complex plots such as subplots, heatmaps, and contour plots.

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.scatter(x,y)
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.title("This is my title")
plt.show()
```

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.

The subplot() function in Matplotlib is used to create a figure with multiple subplots arranged in a grid. It allows for multiple plots to be displayed in a single figure, making it easier to compare different datasets or visualize multiple aspects of the same data.

The subplot() function takes three arguments - nrows, ncols, and index. The nrows and ncols arguments specify the number of rows and columns of subplots in the grid, while the index argument specifies the position of the subplot within the grid.

For example, subplot(2, 2, 1) will create a 2x2 grid of subplots and set the current subplot to the first position in the grid. Subsequent plotting commands will be applied to this subplot until a new subplot is specified.

The subplot() function can be particularly useful when comparing different aspects of the same data or when visualizing multiple datasets side-by-side. It also allows for finer control over the layout and arrangement of plots within a figure.

Overall, the subplot() function is a powerful tool for creating complex and informative visualizations with Matplotlib.

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for each line plot
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create the subplots
fig, axs = plt.subplots(2, 2)

# Plot each line on a separate subplot
axs[0, 0].plot(x1, y1)
axs[0, 1].plot(x2, y2)
axs[1, 0].plot(x3, y3)
axs[1, 1].plot(x4, y4)

# Set the titles for each subplot
axs[0, 0].set_title('Line 1')
axs[0, 1].set_title('Line 2')
axs[1, 0].set_title('Line 3')
axs[1, 1].set_title('Line 4')

# Display the figure
plt.show()
```

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot is a graphical representation of data that uses rectangular bars to represent the magnitudes of different values. In a bar plot, the height or length of each bar corresponds to the value it represents, and the bars are typically arranged along a categorical or numerical axis.

Bar plots are used to compare the magnitudes of different values or groups of values, and are particularly effective for displaying data that can be easily divided into distinct categories. They are also useful for highlighting patterns or trends in the data, such as variations in values over time or differences in performance between different groups or regions.

In addition, bar plots can be customized with a variety of visual features, such as different colors or patterns for each bar, labels for the axes and bars, and error bars or confidence intervals to indicate the uncertainty or variability of the data.

Overall, bar plots are a versatile and widely-used tool for visualizing data, and can be used in a variety of contexts to help interpret and communicate the results of statistical analyses or other data-driven investigations.

```python
import matplotlib.pyplot as plt

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

plt.bar(company,profit)
plt.show()
plt.barh(company,profit)
plt.show()
```

### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

A box plot is a graphical representation of a dataset that provides a summary of its distribution by displaying its quartiles and outliers. In a box plot, a rectangular box is drawn to represent the interquartile range (IQR), which contains the middle 50% of the data. The box is divided by a vertical line to mark the median value of the dataset. Whiskers extend from the top and bottom of the box to represent the minimum and maximum values of the dataset, excluding any outliers, which are shown as individual points beyond the whiskers.

Box plots are used to summarize and compare the distributions of different datasets, especially when the data is skewed or contains outliers. They provide a visual summary of the spread, center, and skewness of the data, and can be used to identify potential outliers or asymmetries in the data. Box plots are also useful for detecting differences between groups of data, as they allow easy comparison of the medians and ranges of each group.

Box plots can be customized with a variety of visual features, such as different colors or patterns for each box or whisker, labels for the axes and boxes, and notches or confidence intervals to indicate the variability or uncertainty of the data. Overall, box plots are a valuable tool for visualizing and summarizing data, and can be used in a variety of contexts, including statistical analysis, data exploration, and data communication.





```python
import matplotlib.pyplot as plt
import numpy as np

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

data = [box1, box2]

# Creating axes instance
fig, ax = plt.subplots()

# Creating axes instance
bp = ax.boxplot(data)

# Adding title and axis labels
ax.set_title('Box Plot')
ax.set_xlabel('Data')
ax.set_ylabel('Value')

# Showing plot
plt.show()
```
