## Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.

The distribution of flight prices in the dataset is right-skewed. This means that there are a few very expensive flights, but most flights are relatively inexpensive. The following code will create a histogram to visualize the distribution of flight prices

```python

plt.figure(figsize=(15,16))
# Create a histogram of the flight prices
sns.histplot(df["Price"],kde=True)

# Add a title to the histogram
plt.title("Distribution of Flight Prices")

# Add labels to the x-axis and y-axis
plt.xlabel("Price")
plt.ylabel("Frequency")

# Show the histogram
plt.show()
```

## Q10. How does the rating of apps vary by category? Create a boxplot to compare the ratings of different app categories.

```python
plt.figure(figsize=(10,6))

# Filter the data to only include apps with ratings
df = df[df["Rating"].notna()]

# Create a dictionary to store the ratings for each category
ratings_by_category = {}
for category in df["Category"].unique():
    ratings_by_category[category] = df[df["Category"] == category]["Rating"].tolist()

# Create a boxplot of the ratings by category
plt.boxplot(ratings_by_category.values())

# Add a title to the boxplot
plt.title("Ratings of Apps by Category")

# Add labels to the x-axis and y-axis
plt.xlabel("Category")
plt.ylabel("Rating")

# Show the boxplot
plt.show()
```
